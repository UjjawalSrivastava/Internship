### Q1 https://www.naukri.com/

```python
# create function to fetch all the details and return in list format.
def fetch_job_details_from_naukri_website(values):
    i = 0
    list_value = []
    for value in values:
        if i == 10:
            break
        i = i+1
        list_value.append(value.text)
    return list_value


# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://www.naukri.com/')

# find element of search and input the value
driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input').send_keys('Data Scientist')
sleep(2)

# find location element and input the value
driver.find_element(By.XPATH, "/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input").send_keys('Delhi/NCR')
sleep(2)

# find search button and click it.
driver.find_element(By.XPATH, '//div[@class="qsbSubmit"]').click()
sleep(5)
    
# find filter and checked it.
driver.find_element(By.XPATH, '//label[@for="chk-3-6 Lakhs-ctcFilter-"]').click()
sleep(5)


# job title details
titles = driver.find_elements(By.XPATH, '//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]//div[1]//a[1]')
    
# job location details
job_locations = driver.find_elements(By.XPATH, '//span[@class="locWdth"]')


# company name details
companies_name = driver.find_elements(By.XPATH, '//span[@class=" comp-dtls-wrap"]//a[1]')

# job Experience details
exp_required = driver.find_elements(By.XPATH, '//span[@class="expwdth"]')

# creating list of all the data
data = [titles, job_locations, companies_name, exp_required]

# store the data into dic format
details = {"Job Title" : fetch_job_details_from_naukri_website(data[0]), "Job Location" : fetch_job_details_from_naukri_website(data[1]), "Company Name" : fetch_job_details_from_naukri_website(data[2]), "Experience Required" : fetch_job_details_from_naukri_website(data[3])}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```

### Q5 https://www.amazon.in/

```python
# creating the function to fetch the details of products and return it in list format.
def fetching_product_details(values):
    value_count = 0
    list_value = []
    for value in values:
        if value_count == 10:
            break
        value_count += 1
        list_value.append(value.text)
    return list_value

# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://www.amazon.in/')

# find element of search and input the value
driver.find_element(By.XPATH,'//input[@id="twotabsearchtextbox"]').send_keys('Laptop')
sleep(2)

# find location element and click it
driver.find_element(By.XPATH, '//input[@id="nav-search-submit-button"]').click()
sleep(2)

# find the element and select the filter.
driver.find_element(By.XPATH, '//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a/div/label/i').click()
sleep(2)

# finding the element and fetching the brand name.
brand_name = driver.find_elements(By.XPATH, '//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]')

# finding the element and fetching the rate of product.
elements = driver.find_elements(By.XPATH, '//div[@class="a-row a-size-small"]//span')

value_count = 0
list_value_of_star = []
for value in elements:
    star_list = value.get_attribute("aria-label")
    
    # Check if the text ends with "out of 5 stars"
    if star_list and star_list.endswith("out of 5 stars"):
        list_value_of_star.append(star_list)
        if value_count == 9:
            break
        value_count += 1

# finding the element and fetching the price of product.
prices = driver.find_elements(By.XPATH, '//span[@class="a-price"]')

# creating list of all the data
data = [brand_name, prices]

# store the data into dic format
details = {"Brand Name" : fetching_product_details(data[0]), "Rating" : list_value_of_star, "Price" : fetching_product_details(data[1])}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```

### Q7 https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1

```python
# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')

# Initialize list for details
sno_list = []
list_of_name = []
list_of_born_dead = []
list_of_term_of_office = []
list_of_remark = []

# Locate the table
table = driver.find_element(By.CSS_SELECTOR, 'table')

# Find all rows in the table
rows = table.find_elements(By.TAG_NAME, 'tr')

# Loop through each row
for row in rows:
    
    # Find all columns in the current row
    columns = row.find_elements(By.TAG_NAME, 'td')
    
    # Check if the row contains the expected number of columns
    if len(columns) == 5:
        # Extract text for each column and append to respective lists
        sno_list.append(columns[0].text)
        list_of_name.append(columns[1].text)
        list_of_born_dead.append(columns[2].text)
        list_of_term_of_office.append(columns[3].text)
        list_of_remark.append(columns[4].text)

# creating list of all the data
data = [list_of_name, list_of_born_dead, list_of_term_of_office, list_of_remark]

# store the data into dic format
details = {"Name" : list_of_name, "Bord-Dead" : list_of_born_dead, "Term of office" : list_of_term_of_office, "Remark" : list_of_remark}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```

### Q8 https://www.motor1.com/

```python
# creating the function to fetch the details of products and return it in list format.
def get_50_most_expensive_car_details(values):
    value_count = 0
    list_value = []
    for value in values:
        if value_count == 50:
            break
        
        if "Price: " in str(value.text):
            list_value.append(str(value.text).replace('Price: ', ''))
        else:
            list_value.append(value.text)
        value_count += 1
    return list_value

# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://motor1.com/')

# find element of search and input the value
search_bar = driver.find_element(By.XPATH,'//input[@class="m1-search-panel-input m1-search-form-text"]').send_keys('50 most expensive cars')
sleep(2)

# finding the element and click on search icon.
search_icon = driver.find_element(By.XPATH, '//button[@class="m1-search-panel-button m1-search-form-button-animate icon-search-svg"]').click()
sleep(2)

# find the element and click on it.
driver.find_element(By.XPATH, '//*[@id="page_index_articles_search"]/div[9]/div/div[1]/div/div/div[1]/div/div[1]/h3/a').click()
sleep(2)

# collect all the car names.
car_name = driver.find_elements(By.XPATH, '//h3[@class="subheader"]')

# collect all the price of cars
car_price = driver.find_elements(By.XPATH, "//strong[contains(text(), 'Price: ') or contains(text(), '$')]")

# creating list of all the data
data = [car_name, car_price]

# store the data into dic format
details = {"Name" : get_50_most_expensive_car_details(data[0]), "Price" : get_50_most_expensive_car_details(data[1])}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```
