```python
from bs4 import BeautifulSoup
import requests

# creating function to get the header tags
def get_header_tags(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for tags in soup.find_all(['h1','h2','h3','h4','h5','h6']):
        return tags.text.strip()

tags = get_header_tags('http://wikipedia.org')
print(tags, end="")
```

## Q2 https://www.imdb.com/list/ls091520106/

```python
import pandas as pd
import re
# creating function to get the top rated 100 movie's details.
def get_top_rated_100_movies(url):
    
    page = requests.get(url)
    
    # Initialize lists to store data
    movie_names = []
    ratings = []
    years = []
    
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        
        containers = soup.find_all('div', class_='lister-item mode-detail')

        for container in containers:
            
            # List of Movies
            movie_name = container.find('h3', class_= 'lister-item-header').text.strip()
            movie_name = re.findall(r'\n(.*?)\n', movie_name, re.DOTALL)
            movie_names.append(str(movie_name).replace('[', '').replace(']', ''))
            
            # Rate
            rate = container.find('span', class_= 'ipl-rating-star__rating').text.strip()
            ratings.append(rate)
            
            # Years
            year = container.find('span', class_= 'lister-item-year text-muted unbold').text.strip()
            years.append(str(year).replace('(', '').replace(')', ''))
            
        data = {'Movie Name': movie_names, 'Released Years': years, 'Rating': ratings}
        df = pd.DataFrame(data)
        return df
    else:
        return 'Permission Denied'
        
            

movies = get_top_rated_100_movies('https://www.imdb.com/list/ls091520106/')
movies
```

## Q3 dineout.co.in

```python
# creating function to get Hotel details.
def get_hotel_detials(url):
    
    page = requests.get(url)
    
    # Initialize lists to store data
    restaurant_names = []
    cuisines = []
    locations = []
    ratings = []
    image_urls = []
    
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        restaurant_containers = soup.find_all('div', class_='restnt-card restaurant')
        
        for container in restaurant_containers:

            # Restaurant name
            name = container.find('a', class_='restnt-name ellipsis').text.strip()
            restaurant_names.append(name)
            
            # Cuisine
            cuisine = container.find('span', class_='double-line-ellipsis').text.strip()
            cuisines.append(str(cuisine)[str(cuisine).find("|") :].replace('|', ''))
            
            # Location
            location = container.find('div', class_='restnt-loc ellipsis').text.strip()
            locations.append(location)
            
            # Image URL
            img = container.find('img', class_= 'no-img')
            image_urls.append(img.get('data-src'))
            
            # Ratings
            rating = container.find('div', class_='restnt-rating rating-4 hide').text.strip()
            ratings.append(rating)
            
        # Create a DataFrame to store the scraped data
        data = {
            'Restaurant Name': restaurant_names,
            'Cuisine': cuisines,
            'Location': locations,
            'Ratings': ratings,
            'Image URL': image_urls
        }
        df = pd.DataFrame(data)
        
        return df

    else:
        return 'Permission Denied'
        
            

hotel_detials = get_hotel_detials('https://www.dineout.co.in/lucknow-restaurants')
hotel_detials
```

## Q4 https://presidentofindia.nic.in/former-presidents

```python
# Creating the function to get the name and terms of farmers office.
def get_farmer_presidents_list(url):
    page = requests.get(url)
    
    farmer_presidents_names = []
    tenure = []
    
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        containers = soup.find_all('div', class_= 'views-row')
        
        for container in containers:
            
            # name of farmer presidents
            name = container.find('h3').text.strip()
            farmer_presidents_names.append(name)
            
            # tenure of president
            date = container.find('h5').text.strip()
            tenure.append(date)
            
        data = {'Name Of President' : farmer_presidents_names, 'Tenure of President' : tenure}
        
        # creating DataFrame
        df = pd.DataFrame(data)
        return df

data = get_farmer_presidents_list('https://presidentofindia.nic.in/former-presidents')
data
```
