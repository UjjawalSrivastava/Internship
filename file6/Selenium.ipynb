```python
# Importing all important module
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from time import sleep
```

### Q1

```python
# create function to fetch all the details and return in list format.
def fetch_job_details(values):
    i = 0
    list_value = []
    for value in values:
        if i == 10:
            break
        i = i+1
        list_value.append(value.text)
    return list_value

# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://www.shine.com/')

# find element of search and click
driver.find_element(By.ID,'webSearchBar').click()
sleep(2)

# find the title element and input the value.
driver.find_element(By.XPATH,"//div[@class='position-relative']//input").send_keys('Data Analyst')
sleep(2)

# find location element and click it
driver.find_element(By.ID, "select_container_id_loc").click()

# find and input the value for location
driver.find_element(By.XPATH,"//div[@id='select_container_id_loc']//input").send_keys('Bangalore')
sleep(2)

# find search button and click it.
driver.find_element(By.XPATH, '//button[@class=" btn btn-secondary undefined"]').click()
    
# job title details
titles = driver.find_elements(By.XPATH, '//strong[@class="jobCard_pReplaceH2__xWmHg"]')

# job location details
job_locations = driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')

# company name details
companies_name = driver.find_elements(By.XPATH, '//div[@class="jobCard_jobCard_cName__mYnow"]//span')

# job Experience details
exp_required = driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')

# creating list of all the data
data = [titles, job_locations, companies_name, exp_required]

# store the data into dic format
details = {"Job Title" : fetch_job_details(data[0]), "Job Location" : fetch_job_details(data[1]), "Company Name" : fetch_job_details(data[2]), "Experience Required" : fetch_job_details(data[3])}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```
