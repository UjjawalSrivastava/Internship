```python
# Importing all important module
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
```

### Q1 https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

```python
# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

# find element of search and input the value
table = driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]')
sleep(2)

rank = 1
rank_list = []
name = []
artist = []
upload_date = []
views = []
rows = table.find_elements(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr')
for row_num in range(1, len(rows) + 1):  # Adjust the range based on the expected number of rows
    xpath = f'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr[{row_num}]/td'
    columns = driver.find_elements(By.XPATH, xpath)
    rank_list.append(rank)
    name.append(columns[0].text.split("[")[0])
    artist.append(columns[1].text)
    views.append(columns[2].text)
    upload_date.append(columns[3].text)
    rank += 1

# store the data into dic format
details = {"Rank" : rank_list, "Name" : name, "Artist" : artist, "Views" : views, "Upload Date" : upload_date}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```

### Q2 https://www.bcci.tv/

```python
# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://www.bcci.tv/')

fixture_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="imw-international-men"]/a[2]')))

date_list = []
time_list = []
series_list = []
place_list = []

# find element and click on it
fixture_element.click()
sleep(5)

cards = driver.find_elements(By.XPATH, '//div[@class="match-info"]')
sleep(2)

for card in cards:
    lines = card.text.strip().split('\n')
    
    # Extract the required information
    date = lines[0]
    time = lines[1]
    series = lines[2]
    place = lines[5]
    
    # store values in list
    date_list.append(date)
    time_list.append(time)
    series_list.append(series)
    place_list.append(place)

# store the data into dic format
details = {"Date" : date_list, "Time" : time_list, "Series" : series_list, "Place" : place_list}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```

### Q4 https://github.com/

```python
# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://github.com/')

# fetching the url of trending page.
trending_page = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
url = trending_page.get_attribute('href')

# trending page
driver.get(url)

# initization of lists.
title_list = []
description_list = []
language_list = []

boxs = driver.find_elements(By.XPATH, '//article[@class="Box-row"]')
sleep(2)
for box in boxs:
    lines = box.text.strip().split('\n')[1:]
    
    title = lines[0]
    description = lines[1]
    language = lines[2].split(' ')
    
    title_list.append(title)
    description_list.append(description)
    language_list.append(language[0])

# store the data into dic format
details = {"Repository title" : title_list, "Repository description" : description_list, "Language used " : language_list}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```

### Q5 https://www.billboard.com/

```python
# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://www.billboard.com/')
sleep(3)

# chart menu click.
chart = driver.find_element(By.XPATH, '//*[@id="main-wrapper"]/header/div/div[2]/div/div/div[2]/div[2]/div/div/nav/ul/li[1]/a')
url = chart.get_attribute('href')

# chart page
driver.get(url)
sleep(3)

# hot page
hot_page_url = driver.find_element(By.XPATH, '//*[@id="main-wrapper"]/main/div[2]/div[1]/div[1]/div/div/div[3]/a')
url = hot_page_url.get_attribute('href')


driver.get(url)
sleep(2)

song_name = []
artist_name = []
last_week = []
peak_rank = []
weeks_board = []

boxs = driver.find_elements(By.XPATH, '//div[@class="o-chart-results-list-row-container"]')
for box in boxs:
    lines = box.text.strip().split('\n')
    
    # Song name
    if lines[1] != 'NEW':
        # Song name
        song_name.append(lines[1])
        
        # Artist name
        artist_name.append(lines[2])
    
        # Last week rank
        last_week.append(lines[3])
    
        # Peak rank 
        peak_rank.append(lines[4])
    
        # Weeks on board
        weeks_board.append(lines[5])
    else:
        # Song name
        song_name.append(lines[2])
        
        # Artist name
        artist_name.append(lines[3])
    
        # Last week rank
        last_week.append(lines[4])
    
        # Peak rank 
        peak_rank.append(lines[5])
    
        # Weeks on board
        weeks_board.append(lines[6])

# store the data into dic format
details = {"Song Name" : song_name, "Artist Name" : artist_name, "Last Week Rank" : last_week, "Peak Rank" : peak_rank, "Weeks Board" : weeks_board}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```

### Q6 https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

```python
# initial driver
driver = webdriver.Chrome()

# open chrome
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')
sleep(3)

book_name = []
author_name = []
volumes_sold = []
publisher = []
genre = []

tr = driver.find_elements(By.XPATH, '//*[@id="article-body-blocks"]/div/table/tbody/tr')
count = len(tr)
for i in range(1, count+1):
    
    xpath = f'//*[@id="article-body-blocks"]/div/table/tbody/tr[{i}]//td'
    columns = driver.find_elements(By.XPATH, xpath)
    
    # Book Name
    book_name.append(columns[1].text)
    
    # Author Name
    author_name.append(columns[2].text)
    
    # Volumes Sold  
    volumes_sold.append(columns[3].text)
    
    # Publisher
    publisher.append(columns[4].text)
    
    # Genre    
    genre.append(columns[5].text)

# store the data into dic format
details = {"Book Name" : book_name, "Author Name" : author_name, "Volumes Sold" : volumes_sold, 
           "Publisher" : publisher, "Genre" : genre}

# creating the dataframe and storing all the data into dataframe.
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# printing the DataFrame
df
```

### Q8 https://archive.ics.uci.edu/

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import pandas as pd

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open UCI Machine Learning Repository
driver.get('https://archive.ics.uci.edu/')
sleep(3)

# Prepare lists to store dataset details
dataset_name = []
task_list = []
data_type = []
attribute_type = []
no_of_instances = []
no_of_attribute = []
year_list = []

# Navigate to the dataset page
url = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/div[1]/header/nav/ul/li[1]/a').get_attribute('href')
driver.get(url)

# Expand all dataset sections
expand_all = driver.find_element(By.XPATH, '//div[@class="swap"]')
sleep(2)
expand_all.click()
sleep(2)

# Collect dataset details
boxes = driver.find_elements(By.XPATH, '//div[@class="rounded-box bg-base-100"]')
for box in boxes:
    
    # Data Name
    dataset_name.extend([name.text for name in box.find_elements(By.XPATH, '//a[@class="link-hover link text-xl font-semibold"]')])
    
    # Task list
    task_list.extend([task.text for task in box.find_elements(By.XPATH, '//p[@class="truncate"]')])
    
    # Data Type
    data_type.extend([td.text for td in box.find_elements(By.XPATH, '//tbody[@class="border"]//tr//td[1]')])
    
    # Attribute type
    attribute_type.extend([td.text for td in box.find_elements(By.XPATH, '//tbody[@class="border"]//tr//td[2]')])
    
    # No of instances
    no_of_instances.extend([ins.text.split(" ")[0] for ins in box.find_elements(By.XPATH, '//div/div/div[2]/div/div[3]/span')])
    
    # No of attribute
    no_of_attribute.extend([att.text for att in box.find_elements(By.XPATH, '//div/div/div[2]/div/div[4]/span')])
    
    # Year     
    year_list.extend([year.text for year in box.find_elements(By.XPATH, '//tbody[@class="border"]//tr//td[3]')])

# Store the data in a dictionary
details = {
    "Dataset Name": dataset_name,
    "Task": task_list,
    "Data Type": data_type,
    "Attribute Type": attribute_type,
    "No of Instances": no_of_instances,
    "No of Attribute": no_of_attribute,
    "Year": year_list
}

# Create a DataFrame and store all data
df = pd.DataFrame(details)

# Close the Chrome WebDriver
driver.close()

# display the DataFrame
df
```
