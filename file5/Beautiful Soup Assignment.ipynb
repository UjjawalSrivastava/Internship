## Q1 https://www.imdb.com/list/ls056092300/

```python
# creating the function to fetch the top 100 indian movies data
def get_top_100_movies(url):
    page = requests.get(url)
    
    names = []
    ratings = []
    years = []
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        containers = soup.find_all('div', class_ = 'lister-item mode-detail')
        for container in containers:
            
            # Movie Names
            name = container.find('h3', class_='lister-item-header').text.strip()
            pattern_name = re.findall(r'\n(.*?)\n', name)
            names.append(str(pattern_name).replace('[', '').replace(']', ''))
            
            # Ratings
            rating = container.find('span', class_= 'ipl-rating-star__rating').text.strip()
            ratings.append(rating)
            
            # Years
            year = container.find('span', class_= 'lister-item-year text-muted unbold').text.strip()
            years.append(str(year).replace('(', '').replace(')', ''))
        
        data = {'Name of Movies' : names, 'Rating' : ratings, 'Years' : years}
        df = pd.DataFrame(data=data)
        return df
    
    else:
        return 'Permission Denied'
    
movies = get_top_100_movies('https://www.imdb.com/list/ls056092300/')
movies
```

## Q6 https://www.bewakoof.com/bestseller?sort=popular%20.

```python
# Creating the function to get the details for first 10 products.
def get_details_of_products(url):
    page = requests.get(url)
    
    product_name = []
    product_price= []
    image_url = []
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        container = soup.find_all('div', class_= 'plp-product-card')
        # container
        for box in container:

            name = box.find('h2', class_='clr-shade4 h3-p-name undefined false').text.strip()
            product_name.append(name)

            price = box.find('div', class_= 'discountedPriceText clr-p-black false').text.strip()
            product_price.append(price)

            image = box.find('img').get('src')
            image_url.append(image)
        
        data = {"Name of Product" : product_name, "Price of Product" : product_price, "Image URL" : image_url}
        df = pd.DataFrame(data=data)
        return df
    else:
        return 'Permission Denied'

url = "https://www.bewakoof.com/bestseller?sort=popular%20"
detials = get_details_of_products(url)
detials
```

## Q7 https://www.cnbc.com/world/?region=world

```python
import numpy as np
import datetime
# Creating function to get the news details
def get_news_details(url):
    page = requests.get('https://www.cnbc.com/world/?region=world')
    current_datetime = datetime.datetime.now()
    headings = []
    links = []
    dates = []
    
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        container = soup.find_all('div', class_='RiverPlusCard-cardLeft')
        for box in container:

            # fetching headings
            heading = box.find('a').text.strip()
            if heading:
                headings.append(heading)

            link = box.find('a').get('href')
            if link == '/pro/':
                continue
            links.append(link)

            date = box.find('span', class_='RiverByline-datePublished')
            if date == None:                  
                dates.append('None')
            else:
                dates.append(date.text.strip())       
                
        data = {"Heading" : headings, "Date" : dates,"News Link" : links}
        df = pd.DataFrame(data=data)
        return df
    else:
        return 'Permission Denied'


news = get_news_details('https://www.cnbc.com/world/?region=world')
news
```

## Q8 https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/

```python
# Creating the function to get the data for articals.
def get_data_of_articals(url):
    page = requests.get(url)
    titles = []
    dates = []
    authors = []
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        container = soup.find_all('div', class_= 'article-listing')
        for box in container:
            title = box.find('h2', class_= 'h5 article-title').text.strip()
            titles.append(title)
            
            date = box.find('p', class_='article-date').text.strip()
            dates.append(date)
            
            author = box.find('p', class_='article-authors').text.strip()
            authors.append(author)
        
        data = {"Artical Title": titles, "Date" : dates, "Author Names" : authors}
        df = pd.DataFrame(data=data)
        return df
    else:
        return 'Permission Denied'
artical_details = get_data_of_articals('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/')
artical_details
```
